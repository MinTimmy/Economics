```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

```python
# Marginal Analysis:
  
MC_x = [0,1,2,3,4,5]
MC_y = [0,3,4,5,8,12]

MB_x = [0,1,2,3,4,5]
MB_y = [6,6,6,6,6,6]

t1 = np.arange(0,1,1)
t2 = np.arange(0,14,2)

plt.figure(figsize=(9,3))
plt.suptitle('Marginal Analysis')

# plt.subplot(133)

plt.plot(MC_x, MC_y, 'bo')
plt.plot(MC_x, MC_y, 'b', label = "MC")
plt.plot(MB_x, MB_y, 'ro')
plt.plot(MB_x, MB_y, 'r', label = "MB")

plt.xlabel('number')
plt.ylabel('MC/MB')
plt.legend()
plt.show()
```

```python
# dinner VS all you can eat
  

MB = [x ** 4 / 1000 - 1
    for x in range(-10,0)]
dinner_MC_function = [2
    for x in range(-10,0)]
all_you_can_eat_MC_function=[0
    for x in range(-10,0)]
plt.suptitle('dinner VS all you can eat')
plt.xlabel('number')
plt.ylabel('MB/MC')
plt.plot(MB, label = 'MB')
plt.plot(dinner_MC_function, label = 'dinner MC')
plt.plot(all_you_can_eat_MC_function, label = 'all you can eat MC')
plt.legend()

plt.show()
```
